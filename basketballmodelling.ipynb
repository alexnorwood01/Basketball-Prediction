```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore', category = FutureWarning )
warnings.filterwarnings('ignore', category = DeprecationWarning )
```

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

```python
lr = LogisticRegression()
parameters = {
    'C':[0.001,0.01,0.1,1,10,100,1000]
}
cv = GridSearchCV (lr, parameters,cv=5)
cv.fit(minmax_scaler,y_train.values.ravel())
print_results(cv)
```

```python
import joblib
joblib.dump(cv.best_estimator_,'LR.pkl')
```

```python
import joblib
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier
import warnings 
warnings.filterwarnings('ignore', category = FutureWarning )
warnings.filterwarnings('ignore', category = DeprecationWarning )
```

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

```python
mlp = MLPClassifier ()
para ={
   'hidden_layer_sizes': [(10,), (50,), (100,)], #simple model could be 100,2 add second layer to perceptron 
    'activation': ['relu','tanh','logistic'],
    'learning_rate': ['constant','invscaling','adaptive']
    }
cv = GridSearchCV(mlp, para, cv=5)
cv.fit(minmax_scaler,y_train.values.ravel())# column vector to an array.values.ravel()
print_results(cv)
```
